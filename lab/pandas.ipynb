```python
from datetime import time
```

```python
def addFractDigit(s):
    cnt=0
    in_fract = False
    for i in range(len(s)):
        if s[i] == '.':
            in_fract = True
        elif in_fract:
            cnt+=1
    if cnt > 0 and cnt < 6:
        s+='0'*(6-cnt)
    return s
```

```python
time.fromisoformat(addFractDigit(d['通話時間'][0]))
```

```python
def cnv(s):
    return time.fromisoformat(addFractDigit(s))
```

```python
d['通話時間']=d['通話時間'].map(cnv)
```
